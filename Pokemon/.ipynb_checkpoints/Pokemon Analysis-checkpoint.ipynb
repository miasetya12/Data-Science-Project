## EDA (Exploratory Data Analysis)

```python
plt.figure(figsize=(5, 6))
ax = sns.countplot(x='Generation', data=df, order=df['Generation'].value_counts().index, )
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(20, 15))

# Count plot for 'Age'
t1 = sns.countplot(x='Type 1', data=df, ax=axes[0])
for p in t1.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
axes[0].set_title('Count Plot of Type 1')

# Count plot for 'Age'
t2=sns.countplot(x='Type 2', data=df, ax=axes[1])
for p in t2.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
axes[1].set_title('Count Plot of Type 2')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
